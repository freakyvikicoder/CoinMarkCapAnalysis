```python
import requests as req
import json
```

```python
jsonUrl = 'https://files.coinmarketcap.com/generated/search/quick_search.json'
jsonDataFromResp = req.get(jsonUrl)
```

```python
jsonToCoinsList = json.loads(jsonDataFromResp.text)
```

```python
listOfAllCoinNames = []
for coinName in jsonToCoinsList:
    listOfAllCoinNames.append(coinName['name'])
```

```python
cleanedUpCoinNameList = []
for coinName in listOfAllCoinNames:
    for character in [' ','(',')','[',']','\\','/']:
        if character in coinName:
            if character == ' ':
                coinName = coinName.replace(character,'-')
            else:
                coinName = coinName.replace(character,'')
    
    cleanedUpCoinNameList.append(coinName)            
```

```python
import pandas as pd
```

```python
def getTimeSeriesDframeForCoins(coinName,startDate,endDate):
    coinFetchSkeletonUrl = 'https://coinmarketcap.com/currencies/{coinName}/historical-data/?start={startDate}&end={endDate}'
    formatStringDict = {'coinName':coinName,'startDate':startDate,'endDate':endDate}
    coinFetchUrl = coinFetchSkeletonUrl.format(**formatStringDict)
    getCoinTableList=[]
    try:
        getCoinTableList = pd.read_html(coinFetchUrl,attrs={'class':'table'})
    except:
        'do nothing'
    if(len(getCoinTableList) > 0):
        return(pd.DataFrame(getCoinTableList[0]))
    else:
        return(pd.DataFrame())
```

```python
def buildExcelSheetFromDframe(coinName,coinDataFrame,excelWriterStream):
    coinDataFrame.to_excel(excelWriterStream,sheet_name=coinName)
```

```python
ethDf = getTimeSeriesDframeForCoins(coinName='ethereum',startDate='20170112',endDate='20180112')
```

```python
bitCoindDf = getTimeSeriesDframeForCoins(coinName='bitcoin',startDate='20170112',endDate='20180112')
```

```python
excelWriterSt = pd.ExcelWriter('CoinMarketCap.xlsx',engine='xlsxwriter')
bitCoindDf.to_excel(excelWriterSt,sheet_name='bitcoin')
ethDf.to_excel(excelWriterSt,sheet_name='ethereum')
excelWriterSt.save()
```

```python
import time
import sys
start = time.time()
excelWriterSt = pd.ExcelWriter('CoinMarketCap.xlsx',engine='xlsxwriter')
try:    
    for coinNameIndi in cleanedUpCoinNameList:
        buildExcelSheetFromDframe(coinName=coinNameIndi,
                                  coinDataFrame=getTimeSeriesDframeForCoins(coinNameIndi,'20170112','20180112'),
                                 excelWriterStream=excelWriterSt)
except:
    e = sys.exc_info()[0]
    print(e)
finally:
    excelWriterSt.save()
end = time.time()
print(end-start)
```
